```python
#43 Find the percentage of developers who have a bachelor's degree.

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("survey_results_public.csv")


ecounts = df["EdLevel"].value_counts()


plt.figure(figsize=(10, 5))
plt.bar(ecounts.index, ecounts.values, color='red',width=0.5)
plt.title("Overall distribution of education")
plt.xlabel("Level of Education")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.show()


plt.figure(figsize=(6, 6))
plt.pie(ecounts, labels=ecounts.index, autopct='%1.1f%%',shadow=True)
plt.title("Overall distribution of education")
plt.show()
```

```python
#42
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("survey_results_public.csv")


ecounts = df["JobSeek"].value_counts()


plt.figure(figsize=(10, 5))
plt.barh(ecounts.index, ecounts.values, color='purple',height=0.5)
plt.title("Distribution of job hunting")
plt.xlabel("hunt status")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.show()


plt.figure(figsize=(6, 6))
plt.pie(ecounts, labels=ecounts.index, autopct='%1.1f%%',shadow=True)
plt.title("Distribution of job hunting")
plt.show()
```

```python
#41 Identify the distribution of developer primary job search methods
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("survey_results_public.csv")


ecounts = df["ResumeUpdate"].value_counts()


plt.figure(figsize=(10, 5))
plt.bar(ecounts.index, ecounts.values, color='green',width=0.4)
plt.title("Distribution of job search method")
plt.xlabel("search status")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.show()


plt.figure(figsize=(6, 6))
plt.pie(ecounts, labels=ecounts.index, autopct='%1.1f%%',shadow=True)
plt.title("Distribution of job search method")
plt.show()
```

```python
#44 Calculate the average number of years since developers last participated in a hackathon.
import pandas as pd
import matplotlib.pyplot as plt


years_since_last_hackathon = df['YearsCode']


years_since_last_hackathon = years_since_last_hackathon.dropna()


years_since_last_hackathon = pd.to_numeric(years_since_last_hackathon, errors='coerce')


average_years_since_last_hackathon = years_since_last_hackathon.mean()

print("Average number of years since last hackathon participation:", average_years_since_last_hackathon)

plt.figure(figsize=(10, 5))


plt.subplot(1, 2, 1)
plt.bar(['Average Years'], [average_years_since_last_hackathon], color='red')
plt.title('Average Years Since Last Hackathon Participation')
plt.ylabel('Years')


plt.subplot(1, 2, 2)
plt.pie([average_years_since_last_hackathon], labels=['Average Years'], autopct='%1.1f%%', colors=['hotpink'])
plt.title('Average Years Since Last Hackathon Participation')

plt.tight_layout()
plt.show()
```
