```python
import numpy as np
import matplotlib.pyplot as plt

def solve_and_plot(a, b, c, title, show_plot=True):
    disc = b**2 - 4*a*c
    r1 = (-b + disc**0.5) / (2*a)
    r2 = (-b - disc**0.5) / (2*a)
    print(f"{title} Roots: {r1}, {r2}")
    x = np.linspace(-5, 5, 400)
    y = a*x**2 + b*x + c
    plt.plot(x, y, label=f"{a}xÂ² + {b}x + {c}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.scatter([r1, r2], [0,0], color='red', label=f"Roots ({title})")
    if show_plot:
        plt.legend(); plt.grid(True); plt.title(title); plt.show()



# Stage 4: Multiple sets in ONE graph
with open("multi_coeff.txt", "w") as f:
    f.write("1 -3 2\n2 -4 -6\n1 2 1")

print("\nMultiple Sets from File:")
plt.figure()
with open("multi_coeff.txt") as f:
    for idx, line in enumerate(f, start=1):
        a, b, c = map(float, line.split())
        solve_and_plot(a, b, c, f"Set {idx}", show_plot=False)

plt.title("Multiple Quadratic Sets in One Graph")
plt.legend()
plt.grid(True)
plt.show()

```
